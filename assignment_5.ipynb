```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("datasets/titanic.csv")
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data.drop("Cabin", axis = 1, inplace = True)
```

```python
data["Age"].fillna(data["Age"].median(), inplace = True)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize = (5,5))
sns.kdeplot(data = data, x = "Age", hue = "Sex", palette={"male": "green", "female":"red"})
```

```python
plt.figure(figsize = (50,30))
sns.countplot(data = data, x = "Age", hue = "Sex")
```

```python
sns.boxplot(data["Age"])
```

```python
sns.boxplot(data = data, x = "Sex", y = "Age", palette={"male": "blue", "female" : "red"})
```

```python
sns.boxplot(data = data, x = "Pclass" , y = "Age")
```

```python
sns.violinplot(data = data, x = "Survived", y="Age")
```
