```python
data['College'] = data['College'].fillna(data['College'].mode())
```

```python
sns.histplot(data['Salary'])
```

```python
data['Salary'].fillna(data['Salary'].median(), inplace = True)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
college_wise = data.groupby(data['College'])
```

```python
college_wise.describe()
```

```python
college_wise_mean_salary = college_wise['Salary'].mean()
```
