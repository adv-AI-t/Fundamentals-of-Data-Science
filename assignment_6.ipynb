```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("datasets/Iris.csv")
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data.drop("Id", axis = 1, inplace = True)
```

```python
species = data["Species"].unique()
```

```python
species
```

```python
fig = plt.figure(figsize = (5,5))
sns.histplot(data = data[data["Species"] == species[0]], x = "SepalLengthCm")
```

```python
fig = plt.figure(figsize = (5,5))
sns.histplot(data = data[data["Species"] == species[1]], x = "SepalLengthCm")
```

```python
fig = plt.figure(figsize = (5,5))
sns.histplot(data = data[data["Species"] == species[2]], x = "SepalLengthCm")
```

```python
fig, axes = plt.subplots(1, 3, figsize = (15, 5))
sns.histplot(data = data[data["Species"] == species[0]], x = "SepalLengthCm", ax = axes[0], color = "pink", kde = True)
axes[0].set_title(species[0])
sns.histplot(data = data[data["Species"] == species[1]], x = "SepalLengthCm", ax = axes[1], color = "violet", kde = True)
axes[1].set_title(species[1])
sns.histplot(data = data[data["Species"] == species[2]], x = "SepalLengthCm", ax = axes[2], color = "indigo", kde = True)
axes[2].set_title(species[2])
```
